```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm

from jmetalpy_social.problem import AckleyFn, DeJongFn, RastriginFn, GriewankFn
```

```python
# 3d

step = 0.05
xx, yy = np.meshgrid(np.arange(-5, 5, step), np.arange(-5, 5, step))

fig, axes = plt.subplots(1, 4, subplot_kw={"projection": "3d"}, figsize=(20, 5))

for fn, ax in zip([AckleyFn, DeJongFn, RastriginFn, GriewankFn], axes.flatten()):
    ack = fn(np.stack([xx, yy], axis=-1))
    surf = ax.plot_surface(xx, yy, ack, linewidth=0, antialiased=False, cmap=cm.coolwarm)
    ax.set_title(fn.__name__[:-2])

plt.show()


# 2d

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for fn, ax in zip([AckleyFn, DeJongFn, RastriginFn, GriewankFn], axes.flatten()):
    ack = fn(np.stack([xx, yy], axis=-1))
    heatmap = ax.imshow(ack, cmap=cm.coolwarm, extent=(xx.min(), xx.max(), yy.min(), yy.max()))
    ax.set_title(fn.__name__[:-2])

plt.show()
```
